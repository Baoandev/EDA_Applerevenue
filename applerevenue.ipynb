```python
import numpy as baoan
import pandas as thaian
import matplotlib.pyplot as thaibaoan
import seaborn as tba #visualisation


ba=thaian.read_csv('D:/applerevenue.csv') 
print(ba.shape)
print(ba[0:10])
```

```python
ba.describe()
```

```python
ba.info()
```

```python
ba.transpose()
```

```python
ba.isnull().sum() # dữ liệu này không có null
```

```python
ba['profit'] = ba['profit or not']

ba.drop(columns=['profit or not'], axis=1, inplace=True)
```

```python
ba['year'] = ba['date'].str[-4:]
print(ba[0:5])
ba_model = ba.drop(['date','year','volume'],axis=1)
```

```python
ba_model.head()
```

```python
ba_grouped = thaian.DataFrame(ba.groupby('year')['close'].mean())
ba_grouped.reset_index(inplace=True)
ba_grouped.head()
```

```python
tba.set_style('darkgrid')
```

```python
thaibaoan.pie(ba['profit'].value_counts(),labels=['KHÔNG LỜI','LỜI'])
thaibaoan.title('Profits Comparsion')
thaibaoan.legend() #vẽ biểu đồ lợi nhuận 
```

```python
thaibaoan.figure(figsize=(21,6))
thaibaoan.plot(ba_grouped['year'],ba_grouped['close'],color='black')
thaibaoan.xlabel('NĂM')
thaibaoan.ylabel('GIÁ TRỊ (ĐÔ LA)')
thaibaoan.title('GIÁ TRỊ TRUNG BÌNH CỦA MỖI NĂM')
```

```python
volume_df = thaian.DataFrame(ba.groupby('year')['volume'].mean())
volume_df.head(5)
```

```python
thaibaoan.figure(figsize=(21,6))
thaibaoan.plot(volume_df.volume)
thaibaoan.title('KHỐI LƯỢNG GIAO DỊCH CỔ PHIẾU THEO NĂM')
thaibaoan.xlabel('NĂM')
thaibaoan.ylabel('KHỐI LƯỢNG CỔ PHIẾU(10e9) ')
```

```python
loinhuan_nam = thaian.DataFrame(ba.groupby('year')['profit'].sum())
loinhuan_nam.head(15)
```

```python

loinhuan_nam.max()
```

```python
thaibaoan.figure(figsize=(21,6))
thaibaoan.plot(loinhuan_nam.profit)
thaibaoan.title('SỐ LẦN CỔ PHIẾU CÓ LỢI NHUẬN TRONG NĂM')
thaibaoan.xlabel('NĂM')
thaibaoan.ylabel('SỐ LẦN')
```

```python
thaibaoan.figure(figsize=(15,10))
thaibaoan.scatter(ba.groupby('year')['profit'].mean(),ba_grouped['year'],color='brown')
thaibaoan.xlabel('LỢI NHUẬN(NẾU P>0.5)')
thaibaoan.ylabel('NĂM')
thaibaoan.title('GIÁ TRỊ TRUNG BÌNH CỦA LỢI NHUẬN (P)')
```

```python
tba.set_palette('viridis')
thaibaoan.scatter(x='open',y='close',data=ba)
thaibaoan.xlabel('Open Values')
thaibaoan.ylabel('Final Values')
thaibaoan.title('Open vs Close Values Comparsion')
```
